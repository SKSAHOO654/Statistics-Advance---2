1 . What is hypothesis testing in statistics ?

Hypothesis testing in statistics is a method used to make decisions or inferences about population parameters based on sample data. It's a formal process for testing an assumption (a hypothesis) about a population.

2 . What is the null hypothesis, and how does it differ from the alternative hypothesis ?

- Null Hypothesis (H₀):

 - Definition:
   - The null hypothesis is a statement that there is no effect, no difference, or no relationship between variables. It's often considered the "default" or "status quo" assumption.
   - In simpler terms, it proposes that any observed differences are due to random chance.
 - Purpose:
   - The null hypothesis is what researchers attempt to disprove.
   - Statistical tests are designed to assess the strength of evidence against the null hypothesis.
 - Example:
   - "There is no difference in the average test scores between students who use method A and students who use method B."
   - "A new drug has no effect on blood pressure."

20 . What is an ANOVA test, and what are its assumptions ?

- ANOVA (Analysis of Variance) is a hypothesis test that checks whether at least one group mean is significantly different from the others.

- It does not tell you which group is different — just that not all means are equal.

21 . What are the different types of ANOVA tests ?

- 1. One-Way ANOVA
 - Use when: You're comparing the means of 3 or more independent groups based on one independent variable.

   - Example: Comparing exam scores of students taught by three different teachers.

   -  Hypothesis: All group means are equal vs. at least one is different.

   -  Simple and most commonly used

- 2. Two-Way ANOVA (Factorial ANOVA)
Use when: You're studying the effect of two independent variables (factors) on one dependent variable.

 -  Example: Comparing how teaching method and student gender affect test scores.

   -  It checks:

     - The main effect of each factor

     - The interaction effect between the two factors

 -  More detailed insights than One-Way ANOVA

- 3. Repeated Measures ANOVA
Use when: You're measuring the same subjects multiple times under different conditions or time points.

 - Example: Testing students' performance before, during, and after a training program.

   - Accounts for within-subject variability

   - Great for time-based or longitudinal studies

22 . What is the F-test, and how does it relate to hypothesis testing ?

image.png

**1. F-test for Equality of Variances**
- Use when: You want to check if two populations have equal variances.

 - H₀ (null): The variances are equal

 - H₁ (alt): The variances are not equal

- Often a pre-step before a t-test, to check if equal-variance assumptions hold.

**2. F-test in ANOVA (Analysis of Variance)**
- Use when: You want to compare the means of 3 or more groups.

 - The F-test tells you if at least one group mean is different by comparing:

   - Between-group variance

   - Within-group variance

- If the F-ratio is large, it means group means are likely different.

**3. F-test in Linear Regression**
- Use when: You want to test the overall significance of a regression model.

 - H₀: All regression coefficients are zero (the model doesn’t explain anything)

 - H₁: At least one coefficient is non-zero (the model is useful)

- The F-test checks whether the regression model significantly improves prediction compared to a model with no predictors.
